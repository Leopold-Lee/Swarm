```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.interpolate import CubicSpline
```

```python
def calculate_distance_and_angle(x1, y1, x2, y2):
    distance = math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)
    angle = math.atan2(y2 - y1, x2 - x1)
    angle_degrees = math.degrees(angle)
    return distance, angle_degrees

def calculate_df_distance_and_angle(df):
    distance1_list = []
    angle1_list = []

    distance2_list = []
    angle2_list = []

    distance3_list = []
    angle3_list = []
    # iterate through the rows
    for index, row in df.iterrows():
        x0 = row['robot0_x']
        y0 = row['robot0_y']
        x1 = row['robot1_x']
        y1 = row['robot1_y']
        x2 = row['robot2_x']
        y2 = row['robot2_y']
        x3 = row['robot3_x']
        y3 = row['robot3_y']

        # robot 0 base
        distance1, angle1 = calculate_distance_and_angle(x0, y0, x1, y1)
        distance2, angle2 = calculate_distance_and_angle(x0, y0, x2, y2)
        distance3, angle3 = calculate_distance_and_angle(x0, y0, x3, y3)

        distance1_list.append(distance1)
        angle1_list.append(angle1)

        distance2_list.append(distance2)
        angle2_list.append(angle2)

        distance3_list.append(distance3)
        angle3_list.append(angle3)

    df['distance0_1'] = distance1_list
    df['angle0_1'] = angle1_list

    df['distance0_2'] = distance2_list
    df['angle0_2'] = angle2_list

    df['distance0_3'] = distance3_list
    df['angle0_3'] = angle3_list
        
    distance1_list = []
    angle1_list = []

    distance2_list = []
    angle2_list = []

    distance3_list = []
    angle3_list = []
    # iterate through the rows
    for index, row in df.iterrows():
        x0 = row['robot0_x']
        y0 = row['robot0_y']
        x1 = row['robot1_x']
        y1 = row['robot1_y']
        x2 = row['robot2_x']
        y2 = row['robot2_y']
        x3 = row['robot3_x']
        y3 = row['robot3_y']

        # robot 1 base
        distance1, angle1 = calculate_distance_and_angle(x1, y1, x0, y0)
        distance2, angle2 = calculate_distance_and_angle(x1, y1, x2, y2)
        distance3, angle3 = calculate_distance_and_angle(x1, y1, x3, y3)

        distance1_list.append(distance1)
        angle1_list.append(angle1)

        distance2_list.append(distance2)
        angle2_list.append(angle2)

        distance3_list.append(distance3)
        angle3_list.append(angle3)


    df['distance1_0'] = distance1_list
    df['angle1_0'] = angle1_list

    df['distance1_2'] = distance2_list
    df['angle1_2'] = angle2_list

    df['distance1_3'] = distance3_list
    df['angle1_3'] = angle3_list

    distance1_list = []
    angle1_list = []

    distance2_list = []
    angle2_list = []

    distance3_list = []
    angle3_list = []
    # iterate through the rows
    for index, row in df.iterrows():
        x0 = row['robot0_x']
        y0 = row['robot0_y']
        x1 = row['robot1_x']
        y1 = row['robot1_y']
        x2 = row['robot2_x']
        y2 = row['robot2_y']
        x3 = row['robot3_x']
        y3 = row['robot3_y']

        # robot 2 base
        distance1, angle1 = calculate_distance_and_angle(x2, y2, x0, y0)
        distance2, angle2 = calculate_distance_and_angle(x2, y2, x1, y1)
        distance3, angle3 = calculate_distance_and_angle(x2, y2, x3, y3)

        distance1_list.append(distance1)
        angle1_list.append(angle1)

        distance2_list.append(distance2)
        angle2_list.append(angle2)

        distance3_list.append(distance3)
        angle3_list.append(angle3)

    df['distance2_0'] = distance1_list
    df['angle2_0'] = angle1_list

    df['distance2_1'] = distance2_list
    df['angle2_1'] = angle2_list

    df['distance2_3'] = distance3_list
    df['angle2_3'] = angle3_list

    distance1_list = []
    angle1_list = []

    distance2_list = []
    angle2_list = []

    distance3_list = []
    angle3_list = []
    # iterate through the rows
    for index, row in df.iterrows():
        x0 = row['robot0_x']
        y0 = row['robot0_y']
        x1 = row['robot1_x']
        y1 = row['robot1_y']
        x2 = row['robot2_x']
        y2 = row['robot2_y']
        x3 = row['robot3_x']
        y3 = row['robot3_y']
        # robot 3 base
        distance1, angle1 = calculate_distance_and_angle(x3, y3, x0, y0)
        distance2, angle2 = calculate_distance_and_angle(x3, y3, x1, y1)
        distance3, angle3 = calculate_distance_and_angle(x3, y3, x2, y2)

        distance1_list.append(distance1)
        angle1_list.append(angle1)

        distance2_list.append(distance2)
        angle2_list.append(angle2)

        distance3_list.append(distance3)
        angle3_list.append(angle3)

    df['distance3_0'] = distance1_list
    df['angle3_0'] = angle1_list

    df['distance3_1'] = distance2_list
    df['angle3_1'] = angle2_list

    df['distance3_2'] = distance3_list
    df['angle3_2'] = angle3_list

```

```python
result_df = pd.DataFrame(columns=['probing angle', 'robot', 'distance', 'distance0', 'angle0', 'distance1', 'angle1', 'distance2', 'angle2', 'distance3', 'angle3'])
```

```python
robots_list = ['robot0', 'robot1', 'robot2', 'robot3']
for index, test_df in enumerate(probing_dfs):
    test_length = len(test_df)
    
    for robot in robots_list:
        origin_robot_x_cs = CubicSpline(origin_df.index, origin_df[robot + '_x'])
        origin_robot_y_cs = CubicSpline(origin_df.index, origin_df[robot + '_y'])
        
        test_robot0_x_cs = CubicSpline(test_df.index, test_df[robot + '_x'])
        test_robot0_y_cs = CubicSpline(test_df.index, test_df[robot + '_y'])

        distance_list = []

        for i in range(distant_slots):
            origin_x = origin_robot_x_cs(i * origin_length / distant_slots)
            origin_y = origin_robot_y_cs(i * origin_length / distant_slots)
            test_x = test_robot0_x_cs(i * test_length / distant_slots)
            test_y = test_robot0_y_cs(i * test_length / distant_slots)
            distance = math.sqrt((origin_x - test_x) ** 2 + (origin_y - test_y) ** 2)

            distance_list.append(distance)

        delay_penalty = test_length / origin_length

        composed_delta = sum(distance_list) * delay_penalty / distant_slots
        print(f'probing angle 0.{index * 5:02}pi, {robot}, composed delta {composed_delta}')
        result_df = result_df.append({'probing angle': f'0.{index * 5:02}pi', 'robot': robot, 'distance': composed_delta}, ignore_index=True)
    
    # save the composed delta to a file
    print()
```

```python
# angle and distance
count_index = 0
robot_indexs = ['0', '1', '2', '3']
for index, test_df in enumerate(probing_dfs):
    test_length = len(test_df)

    for robot_index in robot_indexs:
        
        target_robot_indexs = ['0', '1', '2', '3']
        target_robot_indexs.remove(robot_index)
        for target_robot_index in target_robot_indexs:
            origin_robot_distance_cs = CubicSpline(origin_df.index, origin_df['distance' + robot_index + '_' + target_robot_index])
            origin_robot_angle_cs = CubicSpline(origin_df.index, origin_df['angle' + robot_index + '_' + target_robot_index])
            
            test_robot_distance_cs = CubicSpline(test_df.index, test_df['distance' + robot_index + '_' + target_robot_index])
            test_robot0_angle_cs = CubicSpline(test_df.index, test_df['angle' + robot_index + '_' + target_robot_index])

            distance_list = []
            angle_list = []

            for i in range(distant_slots):
                origin_distance = origin_robot_distance_cs(i * origin_length / distant_slots)
                origin_angle = origin_robot_angle_cs(i * origin_length / distant_slots)
                
                test_distance = test_robot_distance_cs(i * test_length / distant_slots)
                test_angle = test_robot0_angle_cs(i * test_length / distant_slots)

                distance_list.append(abs(origin_distance - test_distance))
                angle_list.append(abs(origin_angle - test_angle))

            delay_penalty = test_length / origin_length

            distance_composed_delta = sum(distance_list) * delay_penalty / distant_slots
            angle_composed_delta = sum(angle_list) * delay_penalty / distant_slots


            result_df.loc[count_index, 'distance' + target_robot_index] = distance_composed_delta
            result_df.loc[count_index, 'angle' + target_robot_index] = angle_composed_delta
            
            # print(f'probing angle 0.{index}pi, robot{robot_index}_{target_robot_index}, composed distance delta {distance_composed_delta}')
            # print(f'probing angle 0.{index}pi, robot{robot_index}_{target_robot_index}, composed angle delta {angle_composed_delta}')
    
        count_index += 1
```

```python
df_metadata = pd.DataFrame(columns=['probing_angle', 'robot', 'angle0_max', 'angle0_min' 'angle0_average', 'distance0_max', 'distance0_min', 'distance0_average', \
                                    'angle1_max', 'angle1_min' 'angle1_average', 'distance1_max', 'distance1_min', 'distance1_average', \
                                    'angle2_max', 'angle2_min' 'angle2_average', 'distance2_max', 'distance2_min', 'distance2_average', \
                                    'angle3_max', 'angle3_min' 'angle3_average', 'distance3_max', 'distance3_min', 'distance3_average'])

robot_indexs = ['0', '1', '2', '3']
for index, test_df in enumerate(probing_dfs):
    # create a empty new row in the df_metadata
    
    for robot_index in robot_indexs:
        new_row = {
            'probing_angle': f'0.{index * 5:02}pi',
            'robot': None,
            'angle0_max': None,
            'angle0_min': None,
            'angle0_average': None,
            'distance0_max': None,
            'distance0_min': None,
            'distance0_average': None,
            'angle1_max': None,
            'angle1_min': None,
            'angle1_average': None,
            'distance1_max': None,
            'distance1_min': None,
            'distance1_average': None,
            'angle2_max': None,
            'angle2_min': None,
            'angle2_average': None,
            'distance2_max': None,
            'distance2_min': None,
            'distance2_average': None,
            'angle3_max': None,
            'angle3_min': None,
            'angle3_average': None,
            'distance3_max': None,
            'distance3_min': None,
            'distance3_average': None
        } 
        target_robot_indexs = ['0', '1', '2', '3']
        target_robot_indexs.remove(robot_index)
        for target_robot_index in target_robot_indexs:
            distance_max = test_df['distance' + robot_index + '_' + target_robot_index].max()
            distance_min = test_df['distance' + robot_index + '_' + target_robot_index].min()
            distance_average = test_df['distance' + robot_index + '_' + target_robot_index].mean()
            
            angle_max = test_df['angle' + robot_index + '_' + target_robot_index].max()
            angle_min = test_df['angle' + robot_index + '_' + target_robot_index].min()
            angle_average = test_df['angle' + robot_index + '_' + target_robot_index].mean()

            new_row['robot'] = 'robot' + robot_index
            new_row['angle' + target_robot_index + '_max'] = angle_max
            new_row['angle' + target_robot_index + '_min'] = angle_min
            new_row['angle' + target_robot_index + '_average'] = angle_average
            
            new_row['distance' + target_robot_index + '_max'] = distance_max
            new_row['distance' + target_robot_index + '_min'] = distance_min
            new_row['distance' + target_robot_index + '_average'] = distance_average
    
    
        df_metadata = df_metadata.append(new_row, ignore_index=True)
```
