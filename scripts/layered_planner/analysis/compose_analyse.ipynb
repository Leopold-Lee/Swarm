```python
import pandas as pd
import math
import matplotlib.pyplot as plt
```

```python
dfs = []
df = pd.read_csv('../result/test0_composed_delta.csv')
df1 = pd.read_csv('../result/test1_composed_delta.csv')
df2 = pd.read_csv('../result/test2_composed_delta.csv')
df3 = pd.read_csv('../result/test3_composed_delta.csv')

dfs.append(df)
dfs.append(df1)
dfs.append(df2)
dfs.append(df3)
```

```python
for df in dfs:
    grouped = df.groupby('robot')
    for name, group in grouped:
        angle_list_float = [float(angle.replace('pi', '')) for angle in group['probing angle']]
        plt.plot(angle_list_float, group['distance'], label=name)

    plt.xlabel('Probing angle (rad)')
    plt.legend()
    plt.show()


# print('=========================')

# for df in dfs:
#     grouped = df.groupby('robot')
#     for name, group in grouped:
#         angle_list_float = [float(angle.replace('pi', '')) for angle in group['probing angle']]
#         plt.plot(angle_list_float, group['distance0'], label=name)

#     plt.xlabel('Probing angle (rad)')
#     plt.legend()
#     plt.show()


# print('=========================')

# for df in dfs:
#     grouped = df.groupby('robot')
#     for name, group in grouped:
#         angle_list_float = [float(angle.replace('pi', '')) for angle in group['probing angle']]
#         plt.plot(angle_list_float, group['distance1'], label=name)

#     plt.xlabel('Probing angle (rad)')
#     plt.legend()
#     plt.show()

# print('=========================')

# for df in dfs:
#     grouped = df.groupby('robot')
#     for name, group in grouped:
#         angle_list_float = [float(angle.replace('pi', '')) for angle in group['probing angle']]
#         plt.plot(angle_list_float, group['distance2'], label=name)

#     plt.xlabel('Probing angle (rad)')
#     plt.legend()
#     plt.show()
```

```python
df0 = pd.read_csv('../result/test0_formation.csv')
df1 = pd.read_csv('../result/test1_formation.csv')
df2 = pd.read_csv('../result/test2_formation.csv')
df3 = pd.read_csv('../result/test3_formation.csv')

dfs = [df0, df1, df2, df3]
```

```python
keywords0 = ['angle1_max', 'angle1_min', 'angle1_average']
keywords1 = ['angle2_max', 'angle2_min', 'angle2_average']
keywords2 = ['angle3_max', 'angle3_min', 'angle3_average']

keywords_list = [keywords0, keywords1, keywords2]

for df in dfs:
    df = df[df['robot'] == 'robot0']
    angle_list_float = [float(angle.replace('pi', '')) for angle in df['probing_angle']]
    # create 1 * 3 subplots
    fig, axes = plt.subplots(1, 3, figsize=(15, 4))

    for i, keyword in enumerate(keywords_list):
        for key in keyword:
            axes[i].plot(angle_list_float, df[key], label= key)
            axes[i].legend()
            axes[i].set_xlabel('Probing angle (rad)')

    plt.show()
```

```python
keywords0 = ['distance1_max', 'distance1_min', 'distance1_average']
keywords1 = ['distance2_max', 'distance2_min', 'distance2_average']
keywords2 = ['distance3_max', 'distance3_min', 'distance3_average']

keywords_list = [keywords0, keywords1, keywords2]

for df in dfs:
    df = df[df['robot'] == 'robot0']
    angle_list_float = [float(angle.replace('pi', '')) for angle in df['probing_angle']]
    # create 1 * 3 subplots
    fig, axes = plt.subplots(1, 3, figsize=(15, 4))

    for i, keyword in enumerate(keywords_list):
        for key in keyword:
            axes[i].plot(angle_list_float, df[key], label= key)
            axes[i].legend()
            axes[i].set_xlabel('Probing angle (rad)')

    plt.show()
```
